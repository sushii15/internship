47. Circular Bar Plot: Display data in a polar/circular bar chart.

```python
import numpy as np
import matplotlib.pyplot as plt

labels = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
values = [50, 70, 65, 90, 30, 55, 80, 60]
angles = np.linspace(0, 2 * np.pi, len(values), endpoint=False) # angles for each label, evenly spaced around the circle , 
# endpoint=False means that the last angle is not included, so it does not overlap with the first angle
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'}) # creates a polar subplot where the angles are used as the x-axis and the values as the y-axis
ax.bar(angles, values, width=0.5, color='skyblue')
ax.set_xticks(angles) # sets the ticks (labels) for the angles
ax.set_xticklabels(labels) # labels for each angle
plt.title("Simple Circular Bar Plot")
plt.show()
```

54. Radar Chart: Create a radar chart to compare different features of a dataset.

```python
labels = ['Speed', 'Power', 'Agility', 'Stamina', 'Skill']
stats = [80, 70, 90, 60, 75]
angles = np.linspace(0, 2 * np.pi, len(labels), endpoint=False).tolist() # means that the angles are evenly spaced around the circle
stats += stats[:1] # to close the radar chart, we add the first value to the end of the list
angles += angles[:1] 
fig, ax = plt.subplots(subplot_kw={'polar': True})
ax.plot(angles, stats)
ax.fill(angles, stats, alpha=0.25) # alpha is for transparency
ax.set_xticks(angles[:-1])  # labels everything except the last angle
ax.set_xticklabels(labels)
plt.title("Radar Chart")
plt.show()
```

56. Animated Line Graph: Animate a live stock market trend line graph.

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np

# Setup the figure and axis
fig, ax = plt.subplots()
x_data, y_data = [], []
line, = ax.plot([], [], color='blue')

# Set axis limits
ax.set_xlim(0, 100)
ax.set_ylim(80, 120)
ax.set_title("Live Stock Market Trend")
ax.set_xlabel("Time")
ax.set_ylabel("Stock Price")

# This function updates the data
def update(frame):
    x_data.append(frame)
    y_data.append(100 + np.random.randn() * 2)  # Simulated stock price
    line.set_data(x_data, y_data)
    ax.set_xlim(0, max(100, frame + 10))        # Expand x-axis if needed
    return line,

# Animate: call update every 100ms
ani = animation.FuncAnimation(fig, update, frames=np.arange(0, 100), blit=True, interval=100)

plt.show()

```

```python


import plotly.graph_objects as go

# Example data: relationship strength between departments
labels = ['Legal', 'Finance', 'Sales', 'HR']

# Define source-target relationships and their strength
sources = [0, 0, 1, 2, 2, 3] 
targets = [1, 2, 2, 0, 3, 0]
values  = [8, 4, 2, 6, 3, 1] 

# Create Sankey-style chord diagram
fig = go.Figure(data=[go.Sankey( # Using Sankey to represent chord diagram
    arrangement = "fixed",
    node=dict(  # nodes in the diagram
        pad=0,
        thickness=20,
        label=labels,
        color="skyblue"
    ),
    link=dict(  # source-target relationships
        source=sources,
        target=targets,
        value=values,
        color="lightgrey"
    )
)])

fig.update_layout(title_text="Chord Diagram (Simplified as Sankey)", font_size=12)
fig.show()
```

59.Word Cloud from Text Data: Generate a word cloud from a dataset of customer reviews.

```python


from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Example: list of customer reviews
reviews = [
    "I love this product, it works great!",
    "Customer service was excellent.",
    "Totally worth the price.",
    "Could be better, but overall satisfied.",
    "Amazing experience! Fast delivery.",
    "The product quality is top-notch."
]

# Join all reviews into one big string
text = " ".join(reviews)

# Generate the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)  
# width and height are the dimensions of the word cloud image, background_color is the colour of the background
# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear') # interpolation='bilinear' is used to smooth the image
plt.axis('off')
plt.title("Customer Reviews Word Cloud")
plt.show()
```

60. Custom Dashboard Using Matplotlib & Seaborn: Build a small dashboard with multiple plots displaying financial data.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# Sample financial data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
revenue = [10000, 12000, 13000, 9000, 15000, 16000]
expenses = [8000, 8500, 9000, 7000, 9500, 11000]
profit = [r - e for r, e in zip(revenue, expenses)]

df = pd.DataFrame({
    'Month': months,
    'Revenue': revenue,
    'Expenses': expenses,
    'Profit': profit
})

# Create a dashboard with 2x2 layout
fig, axs = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle('Financial Dashboard', fontsize=16)

# Revenue Line Plot
sns.lineplot(ax=axs[0, 0], data=df, x='Month', y='Revenue', marker='o')
axs[0, 0].set_title('Monthly Revenue')

# Expenses Bar Plot
sns.barplot(ax=axs[0, 1], data=df, x='Month', y='Expenses', palette='pastel')
axs[0, 1].set_title('Monthly Expenses')

# Profit Line Plot
sns.lineplot(ax=axs[1, 0], data=df, x='Month', y='Profit', marker='s', color='green')
axs[1, 0].set_title('Monthly Profit')

# Revenue vs Expenses Overlap
axs[1, 1].plot(months, revenue, label='Revenue', marker='o')
axs[1, 1].plot(months, expenses, label='Expenses', marker='x')
axs[1, 1].set_title('Revenue vs Expenses')
axs[1, 1].legend()

plt.tight_layout()
plt.show()
```
